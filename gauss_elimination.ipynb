```python
N = 3
arr = [[2, 1, -1, 8], [-3, -1, 2, -11], [-2, 1, 2, -3]]

def print_arr(arr):
  for i in range(N):
    print(arr[i])

def gaussian_elimination(N, arr):
  for i in range(N):
    for j in range(N, -1, -1):
      arr[i][j] /= arr[i][0+i]

#       üîπ Normalization step ‚Üí ‡¶Ø‡ßá‡¶á row-‡¶§‡ßá ‡¶ï‡¶æ‡¶ú ‡¶ï‡¶∞‡¶õ‡ßã ‡¶§‡¶æ‡¶∞ diagonal element arr[i][i] ‡¶¶‡¶ø‡ßü‡ßá ‡¶∏‡¶æ‡¶∞‡¶ø-‡¶ü‡¶æ‡¶∞ ‡¶∏‡¶¨ element ‡¶≠‡¶æ‡¶ó ‡¶ï‡¶∞‡ßã
# üëâ ‡¶Ø‡ßá‡¶® ‡¶ì‡¶á diagonal ‡¶ú‡¶æ‡ßü‡¶ó‡¶æ‡¶∞ value = 1 ‡¶π‡ßü‡•§

# Example: i=0 ‚Üí ‡¶™‡ßç‡¶∞‡¶•‡¶Æ row [2, 1, -1, 8]
# Diagonal = 2 ‚Üí ‡¶∏‡¶¨ element √∑2 ‚Üí [1, 0.5, -0.5, 4]

    for j in range(i+1, N):
      for k in range(N, -1, -1):
        arr[j][k] -= arr[i][k] * arr[j][0+i]

# üîπ ‡¶®‡¶ø‡¶ö‡ßá‡¶∞ ‡¶∏‡¶¨ row ‡¶•‡ßá‡¶ï‡ßá ‡¶ì‡¶á column-‡¶è‡¶∞ ‡¶Æ‡¶æ‡¶®‡¶ï‡ßá ‡¶∂‡ßÇ‡¶®‡ßç‡¶Ø ‡¶ï‡¶∞‡¶æ ‡¶π‡¶ö‡ßç‡¶õ‡ßá‡•§

# Example:
# Row1 ‡¶è‡¶ñ‡¶® [1, 0.5, -0.5, 4]
# Row2 = [-3, -1, 2, -11], ‡¶™‡ßç‡¶∞‡¶•‡¶Æ ‡¶ï‡¶≤‡¶æ‡¶Æ‡ßá‡¶∞ ‡¶Æ‡¶æ‡¶® -3
# factor = -3 ‚Üí ‡¶®‡¶§‡ßÅ‡¶® Row2 = Row2 ‚Äì (factor √ó Row1) =
# [-3,-1,2,-11] - (-3)*[1,0.5,-0.5,4]
# = [0,0.5,0.5,1]

# Row3 ‡¶è‡¶ï‡¶á‡¶≠‡¶æ‡¶¨‡ßá ‡¶Ü‡¶™‡¶°‡ßá‡¶ü ‡¶π‡¶¨‡ßá‡•§

    print_arr(arr)
    print()

  solve = [0 for j in range(N)]
  for i in range(N-1, -1, -1):

# üîπ ‡¶Ø‡¶ñ‡¶® upper triangular ‡¶∂‡ßá‡¶∑ ‚Üí back-substitution ‡¶∂‡ßÅ‡¶∞‡ßÅ
# üîπ ‡¶®‡¶ø‡¶ö ‡¶•‡ßá‡¶ï‡ßá ‡¶â‡¶™‡¶∞‡ßá‡¶∞ ‡¶¶‡¶ø‡¶ï‡ßá ‡¶ó‡¶ø‡ßü‡ßá x, y, z ‡¶¨‡ßá‡¶∞ ‡¶ï‡¶∞‡¶¨‡ßá
    for j in range(N):
      solve[i] = arr[i][N]
      for k in range(i+1, N):
        solve[i] -= arr[i][k] * solve[k]
    solve[i] /= arr[i][i]

# ‡¶ß‡¶∞‡¶æ ‡¶∂‡ßá‡¶∑ row ‡¶π‡ßü‡ßá‡¶õ‡ßá ‚Üí [0, 0, 1, -1]
# üëâ ‡¶§‡¶æ‡¶π‡¶≤‡ßá equation = 1*z = -1 ‚Üí z = -1

# ‡¶§‡¶æ‡¶∞‡¶™‡¶∞ ‡¶â‡¶™‡¶∞‡ßá‡¶∞ row-‡¶§‡ßá ‡¶¨‡¶∏‡¶ø‡¶Ø‡¶º‡ßá y ‡¶¨‡ßá‡¶∞ ‡¶ï‡¶∞‡¶¨‡ßá, ‡¶è‡¶∞‡¶™‡¶∞ x.
  print(solve)

gaussian_elimination(N, arr)


```
